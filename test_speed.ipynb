```python
import ipywidgets as widgets
from IPython.display import display
from jetbot import Robot

# Initialize robot
robot = Robot()

# Widgets
speed_slider = widgets.FloatSlider(value=0.3, min=0.0, max=0.5, step=0.01, description='Speed')
turn_slider = widgets.FloatSlider(value=0.0, min=-0.5, max=0.5, step=0.01, description='Turn')

move_button = widgets.Button(description='Move', button_style='success')
stop_button = widgets.Button(description='Stop', button_style='danger')

left_motor_label = widgets.Label(value='Left Motor: 0.00')
right_motor_label = widgets.Label(value='Right Motor: 0.00')

# Display UI
display(widgets.VBox([
    speed_slider,
    turn_slider,
    widgets.HBox([move_button, stop_button]),
    left_motor_label,
    right_motor_label
]))

# Handlers
def on_move_clicked(b):
    base_speed = speed_slider.value
    turn = turn_slider.value

    # Add turn to left and subtract from right
    left_speed = base_speed + turn
    right_speed = base_speed - turn

    # Clamp between -1.0 and 1.0
    left_speed = max(min(left_speed, 1.0), -1.0)
    right_speed = max(min(right_speed, 1.0), -1.0)
    

    robot.set_motors(left_speed, right_speed)

    # Update display
    left_motor_label.value = f'Left Motor: {left_speed:.2f}'
    right_motor_label.value = f'Right Motor: {right_speed:.2f}'

def on_stop_clicked(b):
    robot.stop()
    left_motor_label.value = 'Left Motor: 0.00'
    right_motor_label.value = 'Right Motor: 0.00'

# Connect handlers
move_button.on_click(on_move_clicked)
stop_button.on_click(on_stop_clicked)
```
